# Matplotlib & Graph plotting using Pandas


Matplotlib :- It is a plotting library for the python language & its numerical mathematics extensions NumPy . It provides object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter , wxPython , QT orGTK+



Advantages of Matplotlib Library :-



1 . Generally easy to get started for simple plots



2 . Supports for custom labels & texts



3 . Great control of every element in a figure



4 . High quality output in every/many formats



5 . Very customisable in general

```python
# importing matplotlib library to use the plotting functions
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0 , 10)
y = np.arange(11 , 21)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D-Graph')
plt.plot(x , y , 'r*' , linestyle = 'dashed' , linewidth = '3' , markersize = 15)
```

Subplots :-

```python
#plotting for position 1
plt.subplot(2 , 2 , 1)
plt.plot(x , y , 'r' , linestyle = "dashed")

# plotting for position 2
plt.subplot(2 , 2 , 2)
plt.plot(x , y , 'b' , linestyle = "solid")

#plotting for position 3
plt.subplot(2 , 2 , 3)
plt.plot(x , y , 'y' , linestyle = "dashdot")

# plotting for position 4
plt.subplot(2 , 2 , 4)
plt.plot(x , y , 'g' , linestyle = "dotted")
```

Graph for Sine Curve :-

```python
x = np.arange(0 , 4*np.pi , 0.1)
y = np.sin(x)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Function Curve')

plt.plot(x , y , 'b' , linewidth = 2)
```

Graph for CoSine Curve :-

```python
x = np.arange(0 , 5*np.pi , 0.1)
y = np.cos(x)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('CoSine Function Curve')

plt.plot(x , y , 'r' , linewidth = 2)
```

Getting the Sin & Cosine Graph both in a single plane by using Subplotting Method

```python
x = np.arange(0 , 5*np.pi , 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2 ,1 ,1)
plt.plot(x , y_sin , 'b' , linewidth = 2)
plt.title('Sine & CoSine Function Curve')

plt.subplot(2 ,1 ,2)
plt.plot(x , y_cos , 'b' , linewidth = 2)
```

Bar Plots :-

```python
x1 = [2,8,10]
y1 = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]

plt.title("Bar Graph")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.bar(x1 , y1 , color = 'r')
plt.bar(x2 , y2 , color = 'b')
```

Histograms :-

```python
a = np.array([22,87,5,43,56,78,35,89,67,83,27,65,49,62,69])
plt.hist(a , bins = 15)
plt.title("Histogram")
plt.show()
```

Piechart :-

```python
labels = ["Python" , "C++" , "Ruby" , "Java" , "C#"]
sizes = [500,400,300,200,100]
colors = ["gold" , "red" , "blue" , "green" , "yellow"]
explode = [0.1,0,0,0,0]

#plotting

plt.pie(sizes , explode = explode , labels = labels , colors = colors , autopct = '%1.1f%%' , shadow = True)

plt.show()
```
