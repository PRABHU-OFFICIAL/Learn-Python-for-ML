CSV [Comma Separated Values]

```python
from io import StringIO, BytesIO

data = ("Col1 , Col2 , Col3\n"
       "x , y , 1\n"
       "a , b , 2\n"
       "c , d , 3")
```

```python
type(data) # returns string
```

```python
StringIO()
```

StringIO :- Text I/O implementation using an in-memory buffer.

```python
df = pd.read_csv(StringIO(data)) # retuns in the DataFrame format (TABULAR FORMAT)
df.head()
```

```python
df.to_csv("Litu_data.csv") # converts the DataFrame to a csv file 
```
