```python
# IMporting the necessary tools and libraries
import numpy as np
import pandas as pd
from io import StringIO , BytesIO

data = ("a,b,c,d\n"
       "1,2,3,4\n"
       "5,6,7,8\n"
       "9,10,11,12\n")
```

```python
print(data)
```

```python
pd.read_csv(StringIO(data) , dtype = {"b" : int , "c" : float })
```

```python
df = pd.read_csv(StringIO(data))
```

```python
df.head()
```

```python
df["b"][2] # returns the element present in the column = "b" and row-index = "2"
```

Index columns & Training Delimiters

```python
data = ("index,col1,col2,col3,col4,col5\n"
       "0,1,2,3,4,5\n"
       "6,7,8,9,10,11\n"
       "12,13,14,15,16,17\n"
       "18,19,20,21,22,23")
print(data)
```

```python
df = pd.read_csv(StringIO(data))
df.head() 
```

```python
df = pd.read_csv(StringIO(data) , index_col = 0) # Index_col makes the given column the dafault indexed column
df.head()
```

Combining usecols & index_cols

```python
data = ("a,b,c\n"
       "4,apple,bat,\n"
       "8,orange,cow,")
```

```python
pd.read_csv(StringIO(data) , usecols = ["b" , "c"] , index_col = False) # returns the data with default indexing and exracting the values present in column b & c only
```

Quoting & Escaping Characters

```python
data = "a,b\n'hello, \\'Bob \\',nice to meet you' , 5"

pd.read_csv(StringIO(data) , escapechar = "\\")
```
