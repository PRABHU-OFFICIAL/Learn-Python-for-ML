```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

Correlation with Heatmap :- A correlation heatma uses coloured cells , typically in a  monochromatic scale , to show a 2D correlation matrix(table between two discrete dimensions or event types. It is very important in feature section.

Heatmap :- It's a feature of Seabrn which helps us to find the correlation between any data items present in a dataset.

Correlation can be found / valid for integers or floating point numbers ranging & t's value can range between -1 to 1.

```python
df.corr() # returns the correlation between numeric data present in the above table .
```

#To visualize the above correlation we can use the heatmap 
sns.heatmap(df.corr()) # as per the ranging in the scale it shows the correlation-values.

```python
sns.heatmap(df.corr())
```

** The above function is used for Feature Selection in Exploratory Data Analysis

Joinplot :- A joinplot allows to study the relationship between 2 numeric values . The central chart display their correlation . It is usually a scatterplot , a hex bin plot, a 2D histogram or a 2D density plot

Bivariate Analysis :- 

Here the analysis is bivariate so the no of parameters should be 2 , Hence all the graphs below shows the analysis between the "tips" and the "total_bill"

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'scatter') 
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'hex') # shows the graph with hexagonal points
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'hist')
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'kde') # shows the graph with kde(Kernel density estimators)
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'reg') # hows the graph with regression line and probabilitydensity function(PDF)
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill' , data = df , kind = 'resid')
```

Pairplot :- A Pairplot is also known as a scatterplot in which one variable is the same data row is matched with another variable's value. Pairplots are just elaborations on this  , showing all variables paired with all other variables .

```python
sns.pairplot(df)
```

Wth parameters such as sex or smoker we can even show the distribution graph using "heu" keyword

```python
sns.pairplot(df , hue = "sex") # here the orange dot is Male & blue dot is Female
```

Distplot :- It helps to check the distribution of the columns feature

```python
sns.distplot(df["tip"])  # returns a histogram along with the kde graph  ( KDE :- Kernel Density Function)
```

```python
# To remove the kde graph we can use the kde = False
sns.distplot(df["tip"] , kde = False , bins = 10)
```
