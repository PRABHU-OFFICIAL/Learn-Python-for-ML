Data Frames :- It is a combination of rows and columns that will basically show the representation format . (i.e., how the data exactly looks in the files)

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.arange(0 , 30).reshape(5,6) , index = ["Row1" , "Row2" , "Row3" , "Row4" , "Row5"] , columns = ["column1" , "column2" , "column3" , "column4" , "column5" , "column6"])
df.head() # returns the stored data present in that file.
```

Conversion :-

```python
df.to_csv("Litu.csv") # Adds tha data present in the table to the following file "Litu.csv"
```

Accessing the elements (loc and iloc) methods :-

loc :- Location ;
iloc :- indexed location

```python
df.loc["Row1"] # returns all the elements present in that row with column specified
```

```python
type(df.loc["Row1"]) # returns series because of 1 row and multiple columns
```

```python
df.iloc[0:2 , 0:2]
```

```python
type(df.iloc[0:2 , 0:2]) # here the no of rows and cloumns both are greater than 1 so it's a dataframe
```

Converting DataFrames to Arrays :-

```python
df.iloc[: , :].values # converts the following dataframe into an array in the same reshaped form (5 , 6)
```

```python
df.iloc[: , :].values.reshape(15 , 2) # we can also reshape it in the form we want
```

```python
df["column1"].unique() # returns an array consisting of all the unique values present in that column
```
